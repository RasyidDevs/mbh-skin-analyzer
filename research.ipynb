```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import json
import torch
device = "cuda" if torch.cuda.is_available() else "cpu"
print(device)
```

```python
def annotate_wrinkle(img_result):
    img = cv2.imread(img_result)
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    lower = np.array([45, 120, 60])   
    upper = np.array([65, 255, 255])  

    mask = cv2.inRange(hsv, lower, upper)

    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

    annotations = []

    for cnt in contours:
        poly = []
        for p in cnt:
            x, y = p[0]
            poly.append([int(x), int(y)])  

        annotations.append({
            "polyline": poly
        })

    return json.dumps(annotations, indent=2)
```

```python
def overlay_annotations(img_origin, annotations):
    img = cv2.imread(img_origin)

    ann = json.loads(annotations)

    for wr in ann:
        pts = wr["polyline"]
        for i in range(len(pts)-1):
            x1, y1 = pts[i]
            x2, y2 = pts[i+1]
            cv2.line(img, (x1, y1), (x2, y2), (0,255,0), 1)  
    return img
```
